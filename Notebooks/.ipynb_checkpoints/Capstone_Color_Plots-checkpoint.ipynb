```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import wfdb
import ecg_cleaning as c

import sys
sys.path.append('..')
from Models import load_functions as f
```

```python
# File path for high frequency sampling
path = '../data/physionet.org/files/ptb-xl/1.0.3/records500/00000/'
file = path+'00004_hr'

signals, fields = wfdb.rdsamp(file, channels=[1])  # Channels = Leads

samp_signal = signals.copy()

sig_len = 5000
sampling_frequency = 500
time = np.arange(0, sig_len) / sampling_frequency

samp_signal_bl = c.baseline_removal(samp_signal, freq_start=0.1, freq_stop=1.5, signal_samp_freq='high')
samp_signal_pli = c.high_freq_removal(samp_signal_bl, freq_start=40, signal_samp_freq='high')


plt.style.use('seaborn-v0_8-whitegrid')

plt.figure(figsize=(12, 5))
plt.plot(time, samp_signal, c='#e8630a', alpha=0.8, linewidth=1, label='Original')
plt.plot(time, np.real(samp_signal_bl), c='r', alpha=0.5, linewidth=1, label='Baseline Removal')
plt.plot(time, np.real(samp_signal_pli), c='b', alpha=0.5, linewidth=1, label='Baseline + High Frequency Removal')
plt.title("Comparing ECG Signals Through Different Levels of Fourier Transform Cleaning")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (mV)")
#plt.grid(True)
plt.legend()
#plt.tight_layout()
plt.savefig('full_fft_ecg.jpg')
plt.show()
```
