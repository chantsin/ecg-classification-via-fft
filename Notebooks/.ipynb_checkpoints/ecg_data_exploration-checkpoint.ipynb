```python
path = '../data/physionet.org/files/ptb-xl/1.0.3/records500/00000/'

record = wfdb.rdrecord(path+'00001_hr')
wfdb.plot_wfdb(record=record, title='Record 00001_hr from PhysioNet Challenge 2015', figsize=(10,8)) 
display(record.__dict__)
```

```python
wfdb.plot_wfdb()
```

```python
dat_file = path+'00001_hr'
#hea_file = path+'00001_hr.hea'

# Read signal and metadata
signals, fields = wfdb.rdsamp(dat_file, sampfrom=0, channels=[0])  # Change channels=[0] based on your needs
sampling_frequency = fields['fs']  # Sampling frequency

# Print metadata
print("Metadata:")
print(fields)

# Convert to time series
num_samples = len(signals)
duration_seconds = num_samples / sampling_frequency
time = [i / sampling_frequency for i in range(num_samples)]

# Plot the ECG signal
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))
plt.plot(time, signals)
plt.title('ECG Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
